1. Simulate bias–variance tradeoff:
○ Fit models of increasing complexity (e.g., polynomial degree 1, 2, 5)
○ Observe underfitting and overfitting
2. Plot training error vs model complexity
3. Explain:
○ What is bias?
○ What is variance?
○ Where does the optimal model lie?

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression, make_classification
import numpy as np
import matplotlib.pyplot as plt

degrees = [1, 2, 5]
errors = []

X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

for d in degrees:
    model = make_pipeline(PolynomialFeatures(d), LinearRegression())
    model.fit(X_reg, y_reg)
    y_pred = model.predict(X_reg)
    
    mse = mean_squared_error(y_reg, y_pred)
    errors.append(mse)

    plt.scatter(X_reg, y_reg)
    plt.scatter(X_reg, y_pred, color='red')
    plt.title(f"Degree {d}")
    plt.show()
```

```python
plt.plot(degrees, errors, marker='o')
plt.xlabel("Model Complexity (Degree)")
plt.ylabel("Training Error")
plt.title("Bias-Variance Tradeoff")
plt.show()
```

Bias
Error due to simplification
High bias → underfitting

Variance
Error due to sensitivity to data
High variance → overfitting
Optimal Model

Middle point where:
Bias ↓
